## Задача 3
- Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.

_Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями._

```python
corr_matrix = np.round(df.corr(), 2)
corr_matrix
```

```python
upp_mat = np.triu(df.corr())
```

```python
plt.figure(figsize = (15,10))
sns.heatmap(df.corr(), annot=True, cmap = 'viridis', mask = upp_mat);
```

```python
plt.figure(figsize=(12,4))
sns.lmplot(data=df, x="sqft_living", y="price", line_kws={'color': 'red'})
plt.xlabel('Квадратура жилой площади, кв. фт')
plt.ylabel('Стоимость дома')
plt.title('Зависимость стоимости дома от квадратуры жилой площади', loc='left');
```

**Выводы**
- согласно проведенному анализу, максимальная корреляция стоимости дома достигает 70% с параметром квадратуры жилой площади;
- основная масса проданных домов имеет жилую площадь до 8 тысяч квадратных футов и стоимость не более 500 тысяч долларов.

```python
plt.figure(figsize=(12, 12))
sns.boxplot(data = df, x = 'grade', y = 'price')
plt.xlabel('Класс дома')
plt.ylabel('Стоимость дома')
plt.title('Зависимость стоимости дома от его класса');
```

**Выводы**
- корреляция класса дома и его стоимости составляет 67%, что является вторым по значению параметром после квадратуры жилой площади;
- классность домов представлена 13 категориями от первого класса (самые дешевые дома) до 13 класса(наиболее дорогие дома)
- основная масса проданных домов имеет жилую площадь до 8 тысяч квадратных футов и стоимость не более 500 тысяч долларов;
- минимальная стоимость домов классов 1-9 приблизительно одинакова, при этом средняя медианная и максимальная стоимости растут по экспоненте и достигают наибольших значений стоимости в 13 классе

```python
plt.figure(figsize=size_)
sns.lmplot(data=df, x="sqft_above", y="price", line_kws={'color': 'red'})
plt.xlabel('Квадратура общей площади дома, кв. фт')
plt.ylabel('Стоимость дома')
plt.title('Зависимость стоимости дома от общей квадратуры дома', loc='left'); 
```

**Выводы**
- корреляция показателей стоимости домов и площади под застройку дома (площадь фундамента) также является достаточно высокой (61%);
- подавляющее большинство домов имеет площадь под застройку до 6 000 квадратных футов и стоимость до 400 000 долларов

```python
plt.figure(figsize=(12, 8))
sns.violinplot(data = df, x = 'waterfront', y = 'price')
plt.xlabel('Вид на набережную')
plt.ylabel('Стоимость дома')
plt.title('Распределение стоимости домов в зависимости от наличия вида на набережную');
```

**Выводы**
- учитывая тот факт, что рассматриваемая нами территория находится на побережье и уходит вглубь материика, 99.2% всех домов не имеют вида на набережную, а стоимость таких домов относительно редко превышает стоимость в 250 000 долларов
- при этом стоимость домов на набережной распределена более равномерно, а стоимость таких домов, в основном, варируется в пределах до 600 000 долларов;
- однако учитывая подавляющее большинство домов без вида на набережную, корреляция является очень слабой и составляет 27%.

```python
plt.figure(figsize=size_)
sns.histplot(data = df, x = 'yr_built', y = 'price', bins = 50)
plt.xlabel('Год постройки дома')
plt.ylabel('Стоимость дома')
plt.title('Распределение стоимости домов в зависимости от года его постройки');
```

**Выводы**
- корреляция же параметров стоимости дома и его стоимости практически равна нулю;
- это вызвано множеством факторов, таких как реновация дома, его квадратура, расположение и т.п.
